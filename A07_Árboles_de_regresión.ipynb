```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
obj = pd.read_csv ("Advertising.csv")
obj.head()
```

### Regresión lineal simple

```python
X = obj [["TV", "radio", "newspaper"]]
y = obj ["sales"]
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
r2 = lr.score(X, y)
r2
```

### Árbol de Decisión

```python
tree = DecisionTreeRegressor(random_state=42)
tree.fit (X,y)
```

```python
r2_tree = tree.score(X,y)
r2_tree
```

### Train-test split 50/50

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
# Train-test split con Regresión Lineal
lr.fit(X_train, y_train)

r2_train_tree = lr.score(X_train, y_train)
r2_test_tree = lr.score(X_test, y_test)

print("R2 en Train:", r2_train_tree)
print("R2 en Test:", r2_test_tree)
```

```python
# Train-test split con Árbol de Decisión
tree.fit(X_train, y_train)

r2_train_tree = tree.score(X_train, y_train)
r2_test_tree = tree.score(X_test, y_test)
print("R2 en Train Tree:", r2_train_tree)
print("R2 en Test Tree:", r2_test_tree)
```

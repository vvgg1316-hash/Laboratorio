Utiliza el archivo "Motor Trend Car Road Tests.xlsx" y completa las siguientes actividades:

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from google.colab import files
uploaded = files.upload()


df = pd.read_excel("Motor Trend Car Road Tests.xlsx")
df.head()
```

### 1.1 Realiza una regresión tomando 'mpg' como salida y eliminando la columna 'model'. Considera todos los demás factores como numéricos/ordinales.

* Calcula el R2 e interpreta los signos de los betas.
* Realiza un train-test-split donde se use el 40% de los datos para entrenar. Calcula el R2 de entrenamiento y de prueba.
* Añade regularización L2 con un hiperparámetro lambda decidido por ti. Cambia este valor y compara con varios distintos los R2 de entrenamiento y de prueba.

```python
df = df.drop(columns=["model"])
```

```python
X = df.drop(columns=["mpg"])
y = df["mpg"]
```

```python
lr1 = LinearRegression()
lr1.fit(X, y)

y_pred = lr1.predict(X)

r2_1 = r2_score(y, y_pred)
print("R2:", r2_1)
```

Esto significa que el modelo explica aproximadamente 86.9% de la variabilidad de mpg

```python
print("Intercepto:", lr1.intercept_)
print("Betas:")
for i in range(len(X.columns)):
    print(X.columns[i], ":", lr1.coef_[i])
```

**Interpretación de los betas**

Beta positivo → cuando la variable aumenta, mpg aumenta

Beta negativo → cuando la variable aumenta, mpg disminuye

**Train-test-split donde se use el 40% de los datos para entrenar**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.40, random_state=42)

lr_split = LinearRegression()
lr_split.fit(X_train, y_train)
r2_train = lr_split.score(X_train, y_train)
r2_test = lr_split.score(X_test, y_test)

print("R2 entrenamiento:", r2_train)
print("R2 prueba:", r2_test)
```

El modelo tiene un R2 muy alto en entrenamiento, pero muy bajo en prueba. Esto significa que el modelo aprendió demasiado bien los datos de entrenamiento, pero no funciona bien con datos nuevos. Esto es un sobreajuste.

**Regularización L2**

```python
ridge_01 = Ridge(alpha=0.1)
ridge_01.fit(X_train, y_train)

print("Lambda = 0.1")
print("R2 train:", ridge_01.score(X_train, y_train))
print("R2 test:", ridge_01.score(X_test, y_test))
```

```python
ridge_1 = Ridge(alpha=1)
ridge_1.fit(X_train, y_train)

print("Lambda = 1")
print("R2 train:", ridge_1.score(X_train, y_train))
print("R2 test:", ridge_1.score(X_test, y_test))
```

```python
ridge_10 = Ridge(alpha=10)
ridge_10.fit(X_train, y_train)

print("Lambda = 10")
print("R2 train:", ridge_10.score(X_train, y_train))
print("R2 test:", ridge_10.score(X_test, y_test))
```

Al agregar regularización L2, el R2 de prueba mejora considerablemente y deja de ser negativo. Esto indica que la regularización ayuda a reducir el sobreajuste. Conforme aumenta lambda, el modelo se vuelve más estable y generaliza mejor.

### 2.1 Realiza una regresión tomando 'mpg' como salida y eliminando la columna 'model'. Crea columnas dummies para los factores 'cyl', 'gear' y 'carb'.

* Calcula el R2 e interpreta los signos de los betas.
* Realiza un train-test-split donde se use el 40% de los datos para entrenar. Calcula el R2 de entrenamiento y de prueba.

```python
df2 = df.copy()
```

```python
df2 = pd.get_dummies(df2, columns=["cyl", "gear", "carb"], drop_first=True)
```

```python
X2 = df2.drop(columns=["mpg"])
y2 = df2["mpg"]
```

```python
lr3 = LinearRegression()
lr3.fit(X2, y2)

y_pred3 = lr3.predict(X2)

r2_3 = r2_score(y2, y_pred3)
print("R2:", r2_3)
```

El modelo con variables dummy obtuvo un R2 de aproximadamente 0.893, lo que indica que explica cerca del 89.3% de la variabilidad de mpg. Este valor es mayor que el modelo anterior sin dummies, por lo que la inclusión de variables categóricas mejora ligeramente el ajuste.

```python
print("Intercepto:", lr3.intercept_)
print("Betas:")
for i in range(len(X2.columns)):
    print(X2.columns[i], ":", lr3.coef_[i])
```

**Train-test-split donde se use el 40% de los datos para entrenar**

```python
X_train3, X_test3, y_train3, y_test3 = train_test_split(X2, y2, train_size=0.40, random_state=42)

lr_split3 = LinearRegression()
lr_split3.fit(X_train3, y_train3)

r2_train3 = lr_split3.score(X_train3, y_train3)
r2_test3 = lr_split3.score(X_test3, y_test3)

print("R2 train:", r2_train3)
print("R2 test:", r2_test3)
```

El modelo con dummies tiene un R2 perfecto en entrenamiento, pero negativo en prueba. Esto significa que el modelo se ajustó demasiado a los datos de entrenamiento y no funciona bien con datos nuevos.

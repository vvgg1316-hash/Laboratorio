```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_excel("TablaAguasFrescas.xlsx")
df.head(9)
```

```python
df["Tamaño"] = df["Tamaño"].replace({
    "Chica": 0,
    "Grande": 1
})
```

```python
df["Hielo"] = df["Hielo"].replace({
    "Sin": 0,
    "Con": 1
})
```

```python
df["Kahlúa"] = (df["Sabor"] == "Kahlúa").astype(int)
df["Horchata"] = (df["Sabor"] == "Horchata").astype(int)
```

```python
df
```

```python
from scipy import stats
import statsmodels.api as sm
```

```python
xf = df[["Tamaño","Hielo","Kahlúa","Horchata"]]
xf1 = sm.add_constant(xf)

y = df['Score'].values.reshape(-1,1)
```

```python
xff = sm.add_constant(xf1)
model = sm.OLS(y,xff)
results = model.fit()
results.summary()
```

## Parte II:

```python
df_1 = df.copy()
```

```python
df_1["Tamaño"] = df_1["Tamaño"].replace({
    0: 1,
    1: -1
    })
```

```python
df_1["Hielo"]  = df_1["Hielo"].replace({
    0: -1,
    1: 1
    })
```

```python
df_1
```

```python
xf_1 = df_1[["Tamaño","Hielo","Kahlúa","Horchata"]]
xf1_1 = sm.add_constant(xf_1)

y_1 = df_1['Score'].values.reshape(-1,1)
```

```python
xff_1 = sm.add_constant(xf1_1)
model_1 = sm.OLS(y_1,xff_1)
results_1 = model_1.fit()
results_1.summary()
```

## Parte III:

**Conclusión:**

Al comparar los resultados de la Parte I y la Parte II, se observa que cambiaron los valores del intercepto y algunos coeficientes. Esto ocurre porque al modificar los números con los que representamos las categorías, también cambia el punto de referencia del modelo.

Sin embargo, el valor de R2 se mantiene igual (0.837), al igual que el F-statistic y los demás indicadores generales. Esto significa que el modelo sigue explicando lo mismo y tiene la misma capacidad para ajustarse a los datos.

En conclusión, los cambios en los coeficientes se deben únicamente a la forma en que se codificaron las variables, pero el modelo en esencia es el mismo.

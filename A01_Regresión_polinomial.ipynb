# Regresión Baterías

### 1er Orden

```python
x = bateria["carga"].values
y = bateria["duracion"].values
```

```python
X1 = x.reshape(-1, 1)
```

```python
lr1 = LinearRegression()
lr1.fit(X1,y)
```

```python
lr1.intercept_
```

```python
lr1.coef_
```

```python
b0 = lr1.intercept_
b1 = lr1.coef_[0]
print(f"La ecuación de la regresión es: y = {b0:.2f} + {b1:.2f}x")
```

```python
r2 =lr1.score(X1,y)
print("R² 1er Orden:", r2)
print(f"El modelo explica aproximadamente el {r2*100:.1f}% del comportamiento de los datos.")
```

### 3er Orden

```python
x2 = x**2
x3 = x**3
```

```python
x = x.reshape(-1,1)
x2 = x2.reshape(-1,1)
x3 = x3.reshape(-1,1)
```

```python
X3 = np.hstack([x, x2, x3])
```

```python
lr3 = LinearRegression()
lr3.fit(X3, y)
```

```python
lr3.intercept_
```

```python
lr3.coef_
```

```python
b0 = lr3.intercept_
b1, b2, b3 = lr3.coef_

print(f"La ecuación de la regresión es: y = {b0:.2f} + {b1:.2f}x + {b2:.2f}x^2 + {b3:.2f}x^3")
```

```python
r2_3 = lr3.score(X3, y)

print("R² 3er Orden:", r2_3)
print(f"El modelo explica aproximadamente el {r2_3*100:.1f}% del comportamiento de los datos.")
```

### 5to Orden

```python
x4 = x**4
x5 = x**5
```

```python
x4 = x4.reshape(-1,1)
x5 = x5.reshape(-1,1)
```

```python
X5 = np.hstack([x, x2, x3, x4, x5])
```

```python
lr5 = LinearRegression()
lr5.fit(X5, y)
```

```python
lr5.intercept_
```

```python
lr5.coef_
```

```python
b0 = lr5.intercept_
b1, b2, b3, b4, b5 = lr5.coef_
print(f"La ecuación de la regresión es: y = {b0:.2f} + {b1:.2f}x + {b2:.2f}x^2 + {b3:.2f}x^3 + {b4:.2f}x^4 + {b5:.2f}x^5")
```

```python
r2_5 = lr5.score(X5, y)

print("R² 5to Orden:", r2_5)
print(f"El modelo explica aproximadamente el {r2_5*100:.1f}% del comportamiento de los datos.")
```

### 7mo Orden

```python
x6 = x**6
x7 = x**7
```

```python
x6 = x6.reshape(-1,1)
x7 = x7.reshape(-1,1)
```

```python
X7 = np.hstack([x, x2, x3, x4, x5, x6, x7])
```

```python
lr7 = LinearRegression()
lr7.fit(X7, y)
```

```python
lr7.intercept_
```

```python
lr7.coef_
```

```python
b0 = lr7.intercept_
b1, b2, b3, b4, b5, b6, b7 = lr7.coef_

print(f"La ecuación de la regresión es: y = {b0:.2f} + {b1:.2f}x + {b2:.2f}x^2 + {b3:.2f}x^3 + {b4:.2f}x^4 + {b5:.2f}x^5 + {b6:.2f}x^6 + {b7:.2f}x^7")
```

```python
r2_7 = lr7.score(X7, y)

print("R² 7mo Orden:", r2_7)
print(f"El modelo explica aproximadamente el {r2_7*100:.1f}% del comportamiento de los datos.")
```

### Comparación de R²

```python
print("R² 1er Orden:", r2)
print("R² 3er Orden:", r2_3)
print("R² 5to Orden:", r2_5)
print("R² 7to Orden:", r2_7)
```

Al aumentar el grado del modelo, el valor de R² también aumenta. El modelo de primer orden es el que peor se ajusta a los datos. A partir del tercer orden, los valores de R² ya son muy altos y las mejoras entre modelos son pequeñas

### Gráfica con nube de dispersión

```python
x_plot = np.linspace(x.min(), x.max(), 300).reshape(-1,1)
```

```python
y1 = lr1.predict(x_plot)
y3 = lr3.predict(np.hstack([x_plot, x_plot**2, x_plot**3]))
y5 = lr5.predict(np.hstack([x_plot, x_plot**2, x_plot**3, x_plot**4, x_plot**5]))
y7 = lr7.predict(np.hstack([x_plot, x_plot**2, x_plot**3, x_plot**4, x_plot**5, x_plot**6, x_plot**7]))
```

```python
plt.scatter(x, y, label="Datos originales")
plt.plot(x_plot, y1, label="Grado 1")
plt.plot(x_plot, y3, label="Grado 3")
plt.plot(x_plot, y5, label="Grado 5")
plt.plot(x_plot, y7, label="Grado 7")
plt.xlabel("Carga")
plt.ylabel("Duración")
plt.title("Regresiones polinomiales - Baterías")
plt.legend()
plt.show()
```

En la gráfica se puede ver cómo los modelos de mayor grado siguen mejor la forma de los datos. El modelo lineal es el que menos se ajusta, mientras que a partir del tercer orden el ajuste es mucho más cercano a los datos originales.
